```python
#7. หาจำนวนผู้หญิงต่อผู้ชายที่อยู่บนเรือ

df5 = pd.read_csv('train.csv')
A=df5['Sex'].value_counts()
print(A)
```

```python
#8. หาจำนวนผู้หญิงและผู้ชายที่รอดชีวิต

df6 = pd.read_csv('train.csv')
check = df6.groupby(['Sex'])['Survived']
check.sum()
```

```python
#9. หา % การรอดชีวิตของผู้หญิงและผู้ชาย

R=len(df6['Survived'])
print(R)

N = df6[(df6['Sex'] == 'female') & (df6['Survived'] == 1)]
N2 =N['Survived'].sum()
print("Number of female survived=", N2)

N3 = df6[(df6['Sex'] == 'male') & (df6['Survived'] == 1)]
N4 =N3['Survived'].sum()
print("Number of male survived=", N4)

print("% of female survived =", (N2/R)*100)
print("% of male survived =", (N4/R)*100)
```

```python
#20. แสดงเป็นกลุ่มช่วงอายุ 0-10, 20-30 …. (เพิ่มทีละ10) และหาว่าแต่ละกลุ่มมีอัตราการรอดชีวิตเท่าไหร่ 

R=len(df7['Survived'])  #คนทั้งหมด

N=df7[(df7['Age']>=0) & (df7['Age']<=10) & (df7['Survived'] == 1)] 
N2 =((N['Survived'].sum())/R)*100
print("% of Age group (0-10) survived=", N2)

N3=df7[(df7['Age']>=11) & (df7['Age']<=20) & (df7['Survived'] == 1)] 
N4 =((N3['Survived'].sum())/R)*100
print("% of Age group (11-20) survived=", N4)

N5=df7[(df7['Age']>=21) & (df7['Age']<=30) & (df7['Survived'] == 1)] 
N6 =((N5['Survived'].sum())/R)*100
print("% of Age group (21-30) survived=", N6)

N7=df7[(df7['Age']>=31) & (df7['Age']<=40) & (df7['Survived'] == 1)] 
N8 =((N7['Survived'].sum())/R)*100
print("% of Age group (31-40) survived=", N8)

N9=df7[(df7['Age']>=41) & (df7['Age']<=50) & (df7['Survived'] == 1)] 
N10 =((N9['Survived'].sum())/R)*100
print("% of Age group (41-50) survived=", N10)

N11=df7[(df7['Age']>=51) & (df7['Age']<=60) & (df7['Survived'] == 1)] 
N12 =((N11['Survived'].sum())/R)*100
print("% of Age group (51-60) survived=", N12)

N13=df7[(df7['Age']>=61) & (df7['Age']<=70) & (df7['Survived'] == 1)] 
N14 =((N13['Survived'].sum())/R)*100
print("% of Age group (61-70) survived=", N14)

N15=df7[(df7['Age']>=71) & (df7['Age']<=80) & (df7['Survived'] == 1)] 
N16 =((N15['Survived'].sum())/R)*100
print("% of Age group (71-80) survived=", N16)

N17=df7[(df7['Age']>=81) & (df7['Age']<=90) & (df7['Survived'] == 1)] 
N18 =((N17['Survived'].sum())/R)*100
print("% of Age group (81-90) survived=", N18)
```
